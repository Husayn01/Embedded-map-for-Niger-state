```python
import geopandas as gpd

niger_state = gpd.read_file("niger_state.shp")
```

```python
import folium

# Get the center of Niger State for the map
center = [niger_state.geometry.centroid.y.mean(), niger_state.geometry.centroid.x.mean()]

# Create a Folium map centered on Niger State
m = folium.Map(location=center, zoom_start=6)

# Add the Niger State shapefile to the map with labels
def style_function(feature):
    return {
        'fillColor': 'blue',
        'color': 'black',
        'weight': 1,
        'fillOpacity': 0.5,
    }

# Add GeoJson with popups for labels
folium.GeoJson(
    niger_state,
    style_function=style_function,
    popup=folium.GeoJsonPopup(fields=['NAME_2'], labels=True)  # Change 'your_label_field' to the actual field name for labels
).add_to(m)

# Display the map
m
```
